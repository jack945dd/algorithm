```
def Cut(p,n):
    MAXp = [0]*(n+1)  #長度為n的最貴價格串列
    for i in range(1, n+1):
        Maxtemp =0  #暫存最貴價格
        for j in range(1, i+1):
            Maxtemp = max(Maxtemp, p[j]+MAXp[i-j])  #比較有沒有更貴的切法
            MAXp[i] = Maxtemp #有更貴的就刷新暫存最貴價格
    return MAXp[n]
p=[0,1,5,8,9,10,17,17,20,24,30] #切割價格串列
aa = int(input("請輸入0-10的數字: "))
print("長度為",aa,"的切割最貴價格為",Cut(p,aa),"元")
```
